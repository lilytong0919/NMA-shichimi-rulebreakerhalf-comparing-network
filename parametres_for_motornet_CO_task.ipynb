**catch_trial_perc**

```python
# First, determine for each trial if it is a catch trial (where EventGo_cue is False throughout the trial)
# or if the result for that trial is 'A' (Aborted).

is_catch_trial = anix_dat1_df.groupby(['animal', 'session', 'trial_id']).apply(
    lambda group: (not group['EventGo_cue'].any()) or (group['result'] == 'A').any() or (group['result'] == 'I').any() or (group['result'] == 'F').any()
)

# Convert the result to a DataFrame
df_trials_status = is_catch_trial.reset_index(name='is_catch')

# Count the number of catch trials per animal and session
catch_trial_counts = df_trials_status[df_trials_status['is_catch']].groupby(['animal', 'session']).size().reset_index(name='num_catch_trials')

# Count the total number of trials per animal and session
total_trial_counts = df_trials_status.groupby(['animal', 'session']).size().reset_index(name='total_trials')

# Merge the counts and calculate the percentage
df_catch_trials = pd.merge(total_trial_counts, catch_trial_counts, on=['animal', 'session'], how='left').fillna(0)
df_catch_trials['catch_trial_perc'] = (df_catch_trials['num_catch_trials'] / df_catch_trials['total_trials']) * 100

# Display the results
display(df_catch_trials)
```

### Calculating **`reaching_distance`** and **`start_position`** from `df`

Based on the `motornet_tf.tasks.CentreOutReach` documentation, `reaching_distance` is the distance of the center-out reach, and `start_position` is the origin of the reach. We can derive suitable values for these parameters from the `df2` data using the `cursor_pos_x`, `cursor_pos_y`, and `EventTarget_Onset` columns.

```python
import numpy as np


# Initialize lists to store calculated reaching distances and start positions for each trial
trial_reaching_distances = []
trial_start_positions = []

# Group df2 by animal, session, and trial_id
grouped_df = anix_dat1_df.groupby(['animal', 'session', 'trial_id'])

for (animal, session, trial_id), trial_df in grouped_df:
    # Find the row(s) where EventTarget_Onset is True
    target_onset_rows = trial_df[trial_df['EventTarget_Onset'] == True]

    if not target_onset_rows.empty:
        # Take the first instance of EventTarget_Onset as the start of the reach
        target_onset_row = target_onset_rows.iloc[0]

        # Convert cursor position from mm to meters (1000 mm = 1 m)
        start_x_m = target_onset_row['cursor_pos_x'] /1000
        start_y_m = target_onset_row['cursor_pos_y'] / 1000
        # if cursor position is in metres
        #start_x_m = target_onset_row['cursor_pos_x']
        #start_y_m = target_onset_row['cursor_pos_y']
        trial_start_positions.append((start_x_m, start_y_m))

        # Filter the DataFrame to include only points from (and after) the target onset time
        movement_df = trial_df[trial_df['time_s'] >= target_onset_row['time_s']].copy()

        # Calculate Euclidean distance from the start position for all points in movement_df
        # Only consider points where cursor position data is not NaN
        valid_movement_df = movement_df.dropna(subset=['cursor_pos_x', 'cursor_pos_y'])

        if not valid_movement_df.empty:
            # Convert cursor positions in valid_movement_df from mm to meters before calculating distance
            distances = np.sqrt(
                ((valid_movement_df['cursor_pos_x'] / 1000) - start_x_m)**2 +
                ((valid_movement_df['cursor_pos_y'] / 1000) - start_y_m)**2
            )
           # distances = np.sqrt(
            #    ((valid_movement_df['cursor_pos_x']) - start_x_m)**2 +
             #   ((valid_movement_df['cursor_pos_y']) - start_y_m)**2
            #)

            if not distances.empty:
                max_distance = distances.max()
                trial_reaching_distances.append(max_distance)

# Calculate the average reaching distance
if trial_reaching_distances:
    mean_reaching_distance = np.mean(trial_reaching_distances)
    print(f"\nSuggested 'reaching_distance': {mean_reaching_distance:.4f} meters")
else:
    print("\nCould not calculate 'reaching_distance'. No valid trials found with 'EventTarget_Onset' and cursor data.")

# Calculate the average start position
if trial_start_positions:
    avg_start_x_m = np.mean([pos[0] for pos in trial_start_positions])
    avg_start_y_m = np.mean([pos[1] for pos in trial_start_positions])
    print(f"Suggested 'start_position': [{avg_start_x_m:.4f}, {avg_start_y_m:.4f}] meters")
else:
    print("Could not determine 'start_position'. No valid trials found with 'EventTarget_Onset' and cursor data.")
```

**gocue_range**

```python
# Group by trial and find the start time of each trial
trial_start_times = anix_dat1_df.groupby(['animal', 'session', 'trial_id'])['time_s'].min()

# For each trial that has an EventGo_cue, find the time of its first occurrence
go_cue_onsets = anix_dat1_df[anix_dat1_df['EventGo_cue'] == True].groupby(['animal', 'session', 'trial_id'])['time_s'].min()

# Calculate the relative go cue onset times for trials that have a go cue.
# We use reindex to ensure proper alignment and .dropna() to exclude catch trials (where go_cue_onsets would be NaN).
relative_go_cue_onsets = (go_cue_onsets - trial_start_times).dropna()

# Determine the overall go_cue_range from these relative onset times
if not relative_go_cue_onsets.empty:
    go_cue_range_min = round(relative_go_cue_onsets.min(),2)
    go_cue_range_max = round(relative_go_cue_onsets.max(),2)
    go_cue_range = [go_cue_range_min, go_cue_range_max]
    print(f"Calculated `go_cue_range` from df1 (relative to the start of the trial): {go_cue_range} seconds")
else:
    print("No trials with 'EventGo_cue' found to calculate `go_cue_range`.")
    go_cue_range = None # Assign None if no go cues are found
```

```python
# For each trial that has an EventTarget_Onset, find the time of its first occurrence
target_onsets = anix_dat1_df[anix_dat1_df['EventTarget_Onset'] == True].groupby(['animal', 'session', 'trial_id'])['time_s'].min()

# For each trial that has an EventGo_cue, find the time of its first occurrence
go_cue_onsets = anix_dat1_df[anix_dat1_df['EventGo_cue'] == True].groupby(['animal', 'session', 'trial_id'])['time_s'].min()

# Calculate the relative go cue onset times for trials that have both a target onset and a go cue.
# This measures the time from target onset to go cue.
relative_go_cue_onsets = (go_cue_onsets - target_onsets).dropna()

# Determine the overall go_cue_range from these relative onset times
if not relative_go_cue_onsets.empty:
    go_cue_range_min = round(relative_go_cue_onsets.min(), 2)
    go_cue_range_max = round(relative_go_cue_onsets.max(), 2)
    go_cue_range = [go_cue_range_min, go_cue_range_max]
    print(f"Calculated `go_cue_range` from df1 (relative to EventTarget_Onset): {go_cue_range} seconds")
else:
    print("No trials with both 'EventTarget_Onset' and 'EventGo_cue' found to calculate `go_cue_range`.")
    go_cue_range = None # Assign None if no go cues are found
```
